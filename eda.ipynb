```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('stock_data/AMD.csv')
df.drop([0,1], axis=0, inplace=True)
df['Price'] = pd.to_datetime(df["Price"])
df[["High", "Low", "Open", "Volume", "Adj Close", "Close"]] = df[["High", "Low", "Open", "Volume", "Adj Close", "Close"]] .astype(float)
```

```python
selected_ticks = df["Price"].tolist()[::1000]
print(selected_ticks)
plt.plot(df.index.values, df["High"])
plt.xticks(df.index[::1000],labels=[t.strftime('%D') for t in selected_ticks], rotation=45)
```
